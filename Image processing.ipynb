# Face Detection


```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("face.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
face_cascade = cv2.CascadeClassifier("face.xml")
faces = face_cascade.detectMultiScale(img)
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
plt.figure(figsize=(5,15))
plt.imshow(img)
plt.show()
```

```python
img = cv2.imread("face.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
face_cascade = cv2.CascadeClassifier("face.xml")
eye_cascade = cv2.CascadeClassifier("eye.xml")
faces = face_cascade.detectMultiScale(img)
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
    eyes = eye_cascade.detectMultiScale(img[y:y+h,x:x+w],1.22)
    for ex,ey,ew,eh in eyes:
        cv2.rectangle(img[y:y+h,x:x+w],(ex,ey),(ex+ew,ey+eh),(0,0,225),3)
plt.figure(figsize=(5,15))
plt.imshow(img)
plt.show()
```

# webcam

```python
import cv2
cam = cv2.VideoCapture(0)
while True:
    ret,img = cam.read()
    cv2.imshow("Webcam",img)
    ##if cv2.waitKey(1) == 13
    if cv2.waitKey(1) == ord("s"):
        break
cv2.destroyAllWindows()
cam.release()
```

```python
import cv2
cam = cv2.VideoCapture(0)
face_cascade = cv2.CascadeClassifier("face.xml")
eye_cascade = cv2.CascadeClassifier("eye.xml")
while True:
    ret,img = cam.read()
    faces = face_cascade.detectMultiScale(img)
    for x,y,w,h in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
        eyes = eye_cascade.detectMultiScale(img[y:y+h,x:x+w],1.22)
    for ex,ey,ew,eh in eyes:
        cv2.rectangle(img[y:y+h,x:x+w],(ex,ey),(ex+ew,ey+eh),(0,0,225),3)
    cv2.imshow("Webcam",img)
    ##if cv2.waitKey(1) == 13
    if cv2.waitKey(1) == ord("s"):
        break
cv2.destroyAllWindows()
cam.release()
```

```python
import cv2
import cvzone
from cvzone.SelfiSegmentModule import SelfSegmentation
cam = cv2.VideoCapture(0)
segment = SelfSegmentation()
while True:
    ret,img = cam.read()
    out_img = segment.removeBG(img,(225,0,0),threshold=0)
    cv2.imshow("Original",img)
    cv2.imshow("New",out_img)
    ##if cv2.waitKey(1) == 13
    if cv2.waitKey(1) == ord("s"):
        break
cv2.destroyAllWindows()
cam.release()
```
